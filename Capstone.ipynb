```python
regular_part_3 = pd.read_csv('../NBA_Player_Evolution_2020-2024/regular_season_box_scores_2010_2024_part_3.csv')

#Change season_year and game_date to date-time format
regular_part_3['game_date'] = pd.to_datetime(regular_part_3['game_date'])

##Select the range of dates between 2020 and 2024(need to go with season year bc the season overlaps 2 year. Not selecting any 24-25 data as it will be inomplete)

regular_part_3 = regular_part_3[regular_part_3['game_date']>='2020-03-01']
#regular_part_3['season_year'] = regular_part_3[regular_part_3['season_year'].isin(filtered_years)]


#Select the columns I will be using to make it easier to work with(I can look at overall team ranks as well)
regular_part_3 = regular_part_3[['season_year', 'game_date', 'gameId', 'teamId', 'teamTricode', 'teamName', 'personId', 'personName', 'minutes', 'assists', 'points', 'fieldGoalsPercentage', 'threePointersPercentage', 'plusMinusPoints', 'freeThrowsPercentage']]

regular_part_3['season_year'].unique()
regular_part_3['minutes'].info()
```

```python
regular_season_totals = pd.read_csv('../NBA_Player_Evolution_2020-2024/regular_season_totals_2010_2024.csv')

#Change season_year and game_date to date-time format
regular_season_totals['GAME_DATE'] = pd.to_datetime(regular_season_totals['GAME_DATE'])

##Select the range of dates between 2020 and 2024(need to go with season year bc the season overlaps 2 year. Not selecting any 24-25 data as it will be inomplete)
regular_season_totals[(regular_season_totals['SEASON_YEAR']>='2020-21') & (regular_season_totals['SEASON_YEAR']<='2023-24')]

#Select the columns I will be using to make it easier to work with(I can look at overall team ranks as well)
regular_season_totals[['SEASON_YEAR', 'GAME_DATE', 'GAME_ID', 'TEAM_ID', 'TEAM_NAME', 'FG_PCT', 'FG3_PCT', 'FT_PCT', 'AST', 'AST_RANK', 'PTS', 'PTS_RANK', 'PLUS_MINUS_RANK']]
```

```python
# Combine the 3 regular season stats for easier analysis and remove duplicates
two_regular_season = pd.merge(regular_part_1, regular_part_2, how='outer')
#all_regular_season = all_regular_season.drop_duplicates()
two_regular_season
```

```python
#merge the combined df with the 3rd regular season data
all_regular_season = pd.merge(two_regular_season, regular_part_3, how='outer')

#remove : and the seconds from the mm:ss fill NaN with 0:0 to account for errors and change to string because the object column doesnt allow
all_regular_season['minutes'] = (all_regular_season['minutes'].fillna('0:0').str.split(':').str[0].astype(int))

#filter for players that played 20 or minutes for more consistent data
all_regular_season = all_regular_season[all_regular_season['minutes'] >= 20]

all_regular_season
```

```python
grouped_data = all_regular_season.groupby(['season_year', 'personName','personId']).agg({'minutes': 'mean', 'points': 'mean', 'assists': 'mean','fieldGoalsPercentage': 'mean'}).reset_index()
```

```python
#filter for the year 2019-2020
year_2019_20 = grouped_data[grouped_data['season_year']=='2019-20']
year_2019_20
```
